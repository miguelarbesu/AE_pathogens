# Comparison of disorder categories according to Disopred

```python
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import sys
sys.path.append('../src/')

import aepathdisorder as aepd

%matplotlib inline

%load_ext autoreload
%autoreload 2
```

```python
# Load aggregated disopred data
protdf = aepd.load_multi_tables('../data/disopred_agg-clas/proteomes/*/')
EHECdf = aepd.load_multi_tables('../data/disopred_agg-clas/EHEC_effectors/')
EPECdf = aepd.load_multi_tables('../data/disopred_agg-clas/EPEC_effectors/')
CRdf = aepd.load_multi_tables('../data/disopred_agg-clas/CR_effectors/')
humandf = aepd.load_multi_tables('../data/disopred_agg-clas/human/')
```

```python
# Calculate class frequencies
protclassfreq = aepd.calc_classfreq(protdf)
EHECclassfreq = aepd.calc_classfreq(EHECdf)
EPECclassfreq = aepd.calc_classfreq(EPECdf)
CRclassfreq = aepd.calc_classfreq(CRdf)
humanclassfreq = aepd.calc_classfreq(humandf)
# Summarize all results
summarydf = pd.concat([protclassfreq, EHECclassfreq, EPECclassfreq,CRclassfreq, humanclassfreq], axis=1)
summarydf.columns = ['A/E pathogens proteomes', 'EHEC effectors', 'EPEC effectors', 'CR effectors', 'Human proteome']
summarydf.to_csv('../data/disopred_agg-clas/all_class_fractions_disopred.tsv',
sep='\t', index_label='Class', float_format='%.3f')
```

```python
taxondict = aepd.load_map('../data/maps/reference_taxa.json')
proteomes = protdf.groupby('dataset')

proteomedict = {}

for taxon, df in proteomes:
    taxon = str(taxon)
    type = taxondict[taxon]['type']
    freqs = aepd.calc_classfreq(df)
    proteomedict[type]=freqs

    
```

```python
fig=plt.figure(figsize=(7,3))

ax = fig.add_subplot(111)

ax.barh('EPEC proteome', proteomedict['EPEC']['IDP'], 
        color='C3')
ax.barh('EPEC proteome', proteomedict['EPEC']['PDR'], 
        left = proteomedict['EPEC'].cumsum()['IDP'], 
        color='C3',
        alpha = 0.66)
ax.barh('EPEC proteome', proteomedict['EPEC']['FRAG'], 
        left = proteomedict['EPEC'].cumsum()['PDR'],
        color='C4',
        alpha=0.33)
ax.barh('EPEC proteome', proteomedict['EPEC']['NDP'], 
        left = proteomedict['EPEC'].cumsum()['FRAG'], 
        color='C0',
        alpha = 0.66)
ax.barh('EPEC proteome', proteomedict['EPEC']['ORD'], 
        left = proteomedict['EPEC'].cumsum()['NDP'], 
        color='C0')

ax.barh('EHEC proteome', proteomedict['EHEC']['IDP'], 
        color='C3')
ax.barh('EHEC proteome', proteomedict['EHEC']['PDR'], 
        left = proteomedict['EHEC'].cumsum()['IDP'], 
        color='C3',
        alpha = 0.66)
ax.barh('EHEC proteome', proteomedict['EHEC']['FRAG'], 
        left = proteomedict['EHEC'].cumsum()['PDR'],
        color='C4',
        alpha=0.33)
ax.barh('EHEC proteome', proteomedict['EHEC']['NDP'], 
        left = proteomedict['EHEC'].cumsum()['FRAG'], 
        color='C0',
        alpha = 0.66)
ax.barh('EHEC proteome', proteomedict['EHEC']['ORD'], 
        left = proteomedict['EHEC'].cumsum()['NDP'], 
        color='C0')

ax.barh('CR proteome', proteomedict['CR']['IDP'], 
        color='C3')
ax.barh('CR proteome', proteomedict['CR']['PDR'], 
        left = proteomedict['CR'].cumsum()['IDP'], 
        color='C3',
        alpha = 0.66)
ax.barh('CR proteome', proteomedict['CR']['FRAG'], 
        left = proteomedict['CR'].cumsum()['PDR'],
        color='C4',
        alpha=0.33)
ax.barh('CR proteome', proteomedict['CR']['NDP'], 
        left = proteomedict['CR'].cumsum()['FRAG'], 
        color='C0',
        alpha = 0.66)
ax.barh('CR proteome', proteomedict['CR']['ORD'], 
        left = proteomedict['CR'].cumsum()['NDP'], 
        color='C0')


ax.set_xlim(0,1)
ax.set_xticks(np.arange(0,1.1,0.1))

# ax.plot(angles, paifrequencies, 'o-', linewidth=2)
# ax.fill(angles, paifrequencies, alpha=0.25)


ax.set_title('Accumulated fraction by structural category')
categories = ["IDP", "PDR", "FRAG", "NDP", "ORD"]
ax.legend(categories, loc='upper center', bbox_to_anchor=(0.5, -0.20),
          fancybox=True, shadow=True, ncol=5)

plt.tight_layout()

plt.savefig('../figures/stackedhbar_ae-proteomes_disopred.png',
            dpi=300)
```
